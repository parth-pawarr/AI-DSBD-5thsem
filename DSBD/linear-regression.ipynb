Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 5, 4, 5]
```

```python
mean_x = sum(x) / len(x)
mean_y = sum(y) / len(y)
```

```python
numerator = sum((x[i] - mean_x) * (y[i] - mean_y) for i in range(len(x)))
denominator = sum((x[i] - mean_x) ** 2 for i in range(len(x)))
m = numerator / denominator
```

```python
b = mean_y - m * mean_x
```

```python
print(f"Slope(m): {m}")
print(f"Intercept(c): {b}")
```

```python
def predict(x_val):
    return m * x_val + b
```

```python
x_test = 6
y_pred = predict(x_test)
print(f"Predicted value at x = {x_test}: {y_pred}")
```
